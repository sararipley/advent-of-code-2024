```python
import numpy as np
import pandas as pd
import csv
```

```python
# read input file and create an array for each list

input = open('day1_input', 'r') 
lines = input.readlines()
list1 = []
list2 = [] 
for line in lines:
    p = line.split()
    list1.append(float(p[0]))
    list2.append(float(p[1]))
input.close() 

print(list1)
print(list2)
```

Part 1:

```python
# Sort lists smallest to largest:
sorted_list1 = sorted(list1)
sorted_list2 = sorted(list2)

# take the difference between each pair of elements in the arrays:
diff = []
for i in range(len(sorted_list1)):
    result = abs(sorted_list1[i] - sorted_list2[i])
    diff.append(result)

# take the sum of the differences
sum_diffs = np.sum(diff)

print (sum_diffs)
# ANSWER: 2344935
```

Part 2:

```python
# create a dictionary where the keys are sorted_list1 and the values are the # of times each element appears in sorted_list2:
counts_dict = {}
for i in sorted_list1:
    counts_dict[i] = sorted_list2.count(i)
print(counts_dict)

# convert dictionary to data frame:
counts_list = list(counts_dict.items())
counts_df = pd.DataFrame(counts_list)

# calculate similarity scores
counts_df['sim_scores'] = counts_df[0] * counts_df[1] 

# calculate overall similarity score
overall_sim_score = np.sum(counts_df['sim_scores'])

print(overall_sim_score)

# ANSWER: 27647262
```
